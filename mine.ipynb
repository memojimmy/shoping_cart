```python
df = pd.merge(left=orders , right=customers , how='inner' , on='customer_id')

df = pd.merge(left= df , right=sales , how= 'inner' , on='order_id')

df = pd.merge(left= df , right=products , how= 'inner' , left_on='product_id' ,right_on='product_ID')
```

```python
df.rename(columns={'quantity_x':'quantity','quantity_y' : 'stock'},inplace=True)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['delivery_date'] = pd.to_datetime(df['delivery_date'])
```

```python
df.drop(columns=['country','payment','customer_id','sales_id','zip_code','home_address'] ,axis=1,inplace=True)
```

```python
df.head()
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True ,ignore_index=True)
```

```python
cat_col = df.select_dtypes(include='object').columns
cat_col
```

```python
for col in cat_col:
    print(col)
    print(df[col])
    print(df[col].unique())
    print(df[col].nunique())
    print('\n' , '=' * 100 , '\n')
```

```python
num_col = df.select_dtypes(include='number').columns.drop(['order_id','product_id'])
num_col
```
